```python
import matplotlib.pyplot as plt
import pandas as pd

# Leer el archivo de datos
data = pd.read_csv('Registro1_OpenBCI.txt', skiprows=6)

# Obtener las seales EEG
eeg = data.iloc[:, 1:9].values

# Crear una figura y un conjunto de ejes
fig, ax = plt.subplots()

# Plotea las seales EEG
ax.plot(eeg)

# Establecer etiquetas de los ejes
ax.set_xlabel('Muestras')
ax.set_ylabel('Amplitud (uV)')

# Agregar leyenda para cada canal
ax.legend(['Canal 1', 'Canal 2', 'Canal 3', 'Canal 4', 'Canal 5', 'Canal 6', 'Canal 7', 'Canal 8'])

# Mostrar el grfico
plt.show()

# Obtener las seales
signals = data.iloc[:, 9:17].values

# Crear una figura y un conjunto de ejes para cada seal
fig, ax = plt.subplots(8, 1, figsize=(10, 12), sharex=True)

# Iterar sobre cada seal y plotearla en su respectivo eje
for i in range(8):
    ax[i].plot(signals[:, i])
    ax[i].set_ylabel('Amplitud (uV)')
    ax[i].set_title(f'Seal {i+1}')
    ax[i].legend([f'Seal {i+1}'])

# Establecer etiquetas del eje x
ax[-1].set_xlabel('Muestras')

# Mostrar el grfico
plt.show()
```
